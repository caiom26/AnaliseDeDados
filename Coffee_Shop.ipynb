```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
DF=pd.read_csv('Dados/coffeeshop_data.csv')
DF.sample(6)
```

```python
DF.head(3)
```

```python
DF=DF.drop('id',axis=1)
DF.head(3)
```

```python
DF.describe()
```

```python
DF.info()
```

```python
DF['date_contacted'] = pd.to_datetime(DF['date_contacted'])
DF.head()
```

```python
for col in DF.columns:
    dados=DF[col].value_counts()
    print(dados)
```

```python
#Como todas as perguntas foram feitas no estado de IL, retiraremos a coluna do DF
```

```python
DF=DF.drop('state',axis=1)
DF.head()
```

```python
num_nulos_competitor_satisaction=DF['competitor_satisfaction'].isnull().sum()
print(num_nulos_competitor_satisaction)
```

```python
#Como o número de nulos nessa coluna representa mais ou menos, 7,5% da quantidade de valores, não os retiraremos, ao contrario, colocaremos a mediana dos valores para substituir os nulos. 
```

```python
DF['competitor_satisfaction'] = DF['competitor_satisfaction'].fillna(DF['competitor_satisfaction'].median())
DF.head()
```

```python
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
```

```python
df_encoded=DF.copy()
```

```python
scaler=MinMaxScaler()
num_cols=df_encoded.select_dtypes(include=['float64', 'int64']).columns
df_encoded[num_cols] = scaler.fit_transform(df_encoded[num_cols])
df_encoded.head()
```

```python
#Agora para aplicarmos o Scaler para as datas, utilizaremos uma métrica de recência
data_base=df_encoded['date_contacted'].max()
df_encoded['recency']=data_base-df_encoded['date_contacted']
df_encoded.head()
```

```python
df_encoded.info()
```

```python
df_encoded['recency']=df_encoded['recency'].dt.days
df_encoded.head()
```

```python
df_encoded=df_encoded.drop('date_contacted',axis=1)
```

```python
df_encoded['recency']=scaler.fit_transform(df_encoded[['recency']])
df_encoded.head()
```

```python
bool_cols=df_encoded.select_dtypes(include=['bool']).columns
df_encoded[bool_cols]=df_encoded[bool_cols].astype(int)
df_encoded.head()
```

```python
encoder=OneHotEncoder(sparse=False)

obj_cols=df_encoded.select_dtypes(include=['object']).columns
encoded_cols = encoder.fit_transform(df_encoded[obj_cols])
encoded_col_names = encoder.get_feature_names_out(obj_cols)
df_encoded_onehot = pd.DataFrame(encoded_cols, columns=encoded_col_names)
df_encoded = pd.concat([df_encoded.drop(columns=obj_cols), df_encoded_onehot], axis=1)
df_encoded.head()
```

```python
from sklearn.model_selection import train_test_split
y=df_encoded['bought_coffee']
x=df_encoded.drop('bought_coffee',axis=1)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train, y_train)
rf_predictions = rf_model.predict(x_test)
rf_accuracy = accuracy_score(y_test, rf_predictions)
rf_precision = precision_score(y_test, rf_predictions)
rf_recall = recall_score(y_test, rf_predictions)
rf_f1 = f1_score(y_test, rf_predictions)
print("Random Forest Classifier:")
print(f"Acurácia: {rf_accuracy:.4f}")
print(f"Precisão: {rf_precision:.4f}")
print(f"Recall: {rf_recall:.4f}")
print(f"F1 Score: {rf_f1:.4f}")
```

```python
def is_outlier(array, extreme=False):
  """Custom function to identify outliers in an array"""
  q1,q3 = np.quantile(array, [.25,.75])
  iqr = q3-q1

  factor = 3. if extreme else 1.5
  upper_outlier = q3 + factor*iqr
  lower_outlier = q1 - factor*iqr

  return (array < lower_outlier) | (array > upper_outlier)
```

```python
def string_para_float(valor):
    """
    Converte uma string no formato '12.691,51' para float.

    Parâmetros:
    valor : str
        O valor em string que será convertido para float.

    Retorna:
    float
        O valor convertido para float.
    """
    # Remover o ponto (separador de milhar) e substituir a vírgula por ponto (separador decimal)
    valor = valor.replace('.', '').replace(',', '.')

    # Converter para float
    return float(valor)


```
